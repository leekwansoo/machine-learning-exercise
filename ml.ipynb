# Machine learning  Basic Project

```python
from sklearn import datasets

housing = datasets.fetch_california_housing()

print(housing.DESCR)
```

```python
print(housing.feature_names)
```

```python
import pandas as pd
x = housing.data
y = housing.target
df1 = pd.DataFrame(x, columns=["MedInc", "HouseAge", "AveRooms", "AveBedrms", "Population", "AveOccup", "Latitude", "Longitude"])
df2 = pd.DataFrame(y, columns=['target'])
print(df1.head())
print(df2.head())
```

```python
# Training and Testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42) # split data into training and testing sets
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
# model = KNeighborsRegressor(n_neighbors=5) # create KNN regression model with 5 neighbors
model = LinearRegression() # create linear regression model
model.fit(x_train, y_train) # fit the model to the training data

y_pred = model.predict(x_test)
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)
print("R^2 Score:", r2)
```

```python
# Optimization

from sklearn.preprocessing import StandardScaler, PolynomialFeatures

x = housing.data
y = housing.target

poly = PolynomialFeatures(degree=2)
x = poly.fit_transform(x)
print(x.shape)
model.fit(x, y)
y_pred = model.predict(x)
# Mean Squared Error is a measure of the average squared difference between the predicted values and the actual values.
print("Mean Squared Error:", mean_squared_error(y, y_pred))
# R^2 Score is a measure of how well the model fits the data. 
# It ranges from 0 to 1, where 1 indicates a perfect fit and 0 indicates that the model does not explain any of the variance in the data.
print("R^2 Score:", r2_score(y, y_pred))

```

```python
# Optimization with Ensemble Methods

from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor

LR = LinearRegression()
GBR = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, max_iter =100, random_state=42)
RFR = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1) # use all cpu cores for training

for i in [LR, GBR, RFR]:
    i.fit(x_train, y_train)
    y_pred = i.predict(x_test)
    r2 = r2_score(y_test, y_pred)
    print(f"{i.__class__.__name__} Mean Squared Error:", mean_squared_error(y_test, y_pred))
    print(f"{i.__class__.__name__} R^2 Score:", r2)
```

```python
# Find best GBR parameters using GradientBoostRegressor
# Optimization with Ensemble Methods
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor, HistGradientBoostingRegressor

# GBR = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, max_iter =100, random_state=42)

for j in [0.01, 0.1, 0.2]:
    for i in [200, 250, 350]:
        model = HistGradientBoostingRegressor(learning_rate=j, max_iter = i, random_state=42)
        model.fit(x_train, y_train)
        y_pred = model.predict(x_test)
        r2 = r2_score(y_test, y_pred)
        print(f"{model.__class__.__name__} Mean Squared Error:", mean_squared_error(y_test, y_pred))
        print(f"{model.__class__.__name__}, {i}, R^2 Score:", r2)


```

```python
# save the model using joblib
import joblib
# Find best GBR parameters using HistGradientBoostingRegressor
# Optimization with Ensemble Methods
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor, HistGradientBoostingRegressor

# GBR = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, max_iter =100, random_state=42)

learning_rate = 0.05
max_iter = 200

model = HistGradientBoostingRegressor(
    learning_rate=learning_rate, 
    max_iter = max_iter, 
    random_state=42)
model.fit(x_train, y_train)

joblib.dump(model, 'my_model.joblib')
y_pred = model.predict(x_test)
r2 = r2_score(y_test, y_pred)
# print(f"{model.__class__.__name__} Mean Squared Error:", mean_squared_error(y_test, y_pred))
print(f"{model.__class__.__name__}, {max_iter}, R^2 Score:", r2)


```

```python
# Predict using saved model
loaded_model = joblib.load('my_model.joblib')
y_pred_loaded = loaded_model.predict(x_test)
r2_loaded = r2_score(y_test, y_pred_loaded)
print(f"{loaded_model.__class__.__name__}, {max_iter}, R^2 Score:", r2_loaded)
```
